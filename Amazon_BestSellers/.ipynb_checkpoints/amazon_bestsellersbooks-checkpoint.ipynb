## 4. Preparação dos Dados

```python
# Formatando colunas categóricas:
coluna_categorica = list(amazon.select_dtypes(exclude=('int', 'float')).columns)

print(f'As colunas que possuem valores categóricos: {", ".join(coluna_categorica)}.')
```

### 4.1 Valores Duplicados

```python
# Analisando colunas com valores duplicados:
for coluna in coluna_categorica:
    if amazon[coluna].duplicated().any() == True:
        print (f'A coluna {coluna} contém valores duplicados.')
    else:
        print (f'A coluna {coluna} não contém valores duplicados.')
```

### 4.2 Erros e Correções

Vamos agora verificar se as colunas acima contêm palavras que diferem apenas por letras maiúsculas e minúsculas e por espaços extras.

```python
# Verificando erros de digitação:
for coluna in coluna_categorica:
    print(f'{coluna}(antes): {len(set(amazon[coluna]))}'
          f' | {coluna}(depois): {len(set(amazon[coluna].str.title().str.strip()))}')
```

Assim, foi verificado que houve 1 erro de digitação na coluna _Name_. 

#### Coluna _Name_:

```python
# Corrigindo erro de digitação:
amazon['Name'] = amazon['Name'].str.title().str.strip()
```

```python
# Verificando se os erros de ortografia foram corrigidos:
for coluna in coluna_categorica:
    print(f'{coluna}(antes): {len(set(amazon[coluna]))}'
          f' | {coluna}(depois): {len(set(amazon[coluna].str.title().str.strip()))}')
```

#### Coluna _Author_:

Vamos verificar se há nomes de autores repetidos, porém escritos com grafias diferentes.

```python
# Analisando todos os valores da coluna de autores:
authors = amazon['Author'].sort_values().unique()
#authors
```

Os seguintes nomes de autores:
- George R. R. Martin 
- George R.R. Martin 
- J. K. Rowling 
- J.K. Rowling 

têm grafias diferentes, embora sejam os mesmos autores. 

```python
# Substituindo os nomes dos autores:
amazon = amazon.replace('George R. R. Martin', 'George R.R. Martin')
amazon = amazon.replace('J. K. Rowling', 'J.K. Rowling')
```

```python
# Verificando se as alterações foram bem sucedidas:
for coluna in coluna_categorica:
    print(f'{coluna}(antes): {len(set(amazon[coluna]))}'
          f' | {coluna}(depois): {len(set(amazon[coluna].str.title().str.strip()))}')
```

#### Coluna _Genre_:


```python
# Verificando os valores da coluna de gêneros:
genre = amazon['Genre'].unique()
genre
```

#### Coluna _Year_:

```python
# Verificando a coluna dos anos:
years = list(amazon['Year'].unique())
sorted(years)
```

#### Coluna _User Rating_:

```python
# Verificando a coluna de avaliações dos usuários:
rating = list(amazon['User Rating'].unique())
sorted(rating)
```

```python
# Removendo valores duplicados:
amazon = amazon.drop_duplicates(subset='Name', keep='last')
```

A partir desta primeira análise, foi possível obter as seguintes informações:

## 5. Análise Exploratória - Dados Categóricos

Nesta seção, vamos analisar os dados categóricos para descobrir:

- Quais autores foram mais bem avaliados.
- Quais autores publicaram mais best sellers.
- Quais autores tiveram mais resenhas de livros.
- Quais livros tiveram mais resenhas e consequentemente mais leitores.
- Os livros de quais autores receberam a classificação média mais alta.
- Quais gêneros se tornaram best-sellers com mais frequência.

Vamos criar novos dataframes com base na classificação média dos autores, no número de livros escritos pelos autores, no número de resenhas de livros e no número de livros por gênero. Então faremos a visualização das informações coletadas.

#### Autores mais bem Avaliados:

```python
# Agrupamos todos os autores com a classificação média mais alta:
authors_by_rating = amazon.groupby('Author')[['User Rating']].mean()\
                                            .sort_values('User Rating', ascending=False)\
                                            .reset_index()
#authors_by_rating
```

```python
# Visualização em tabela dos melhores autores:
fig_1 = make_subplots(rows=1, cols=1, subplot_titles=(""))


# Parâmetros da tabela:
fig_1.add_trace(go.Table(columnorder = [1,2],
                         columnwidth = [20,10],
                         header=dict(values=authors_by_rating.columns, 
                                     line_color='darkslategray',
                                     fill_color='Salmon',
                                     height=30),
                         cells=dict(values=[authors_by_rating['Author'][:10],
                                    authors_by_rating['User Rating'][:10]], 
                                    line_color='darkslategray',
                                    fill_color='White')))

# Parâmetros para exibição da tabela:
fig_1.update_layout(showlegend=False, 
                    title_text='Top 10 Autores Avaliados:', 
                    title_font_size=22, 
                    title_font_family='Arial', 
                    title_x=0.5,
                    font=dict(family='Arial', 
                              size=12, 
                              color='black'))

# Exibindo a tabela:
fig_1.show()
```

```python
# top 10 autores mais bem avaliados
plt.figure(figsize = (30,10))

sns.barplot(
    x = authors_by_rating["Author"][:10], 
    y = authors_by_rating["User Rating"][:10],
    palette = 'magma' )

plt.title("Autores com as Melhores Avaliações", fontsize = 30)
plt.xticks(fontsize = 25, rotation = 70)
plt.yticks(fontsize = 20)
plt.xlabel(" ",fontsize = 30)
plt.ylabel("User Rating",fontsize = 25)

plt.show()
```

#### Autores com mais Best Sellers:

```python
# Agrupamos os autores que escreveram mais best-sellers:
authors_by_books_written = amazon.groupby('Author')[['Name']].count()\
                                                                 .sort_values('Name', ascending=False)\
                                                                 .reset_index()
#authors_by_books_written
```

```python
# Autores que escreveram mais best sellers:
plt.figure(figsize = (30,10))

sns.barplot(
    x = authors_by_books_written["Author"][:10], 
    y = authors_by_books_written["Name"][:10],
    palette = 'magma' )

sns.lineplot(
    x = authors_by_books_written["Author"][:10], 
    y = authors_by_books_written["Name"][:10],
    color = "Salmon" )

plt.title("Top 10 Autores Com Mais Best Sellers", fontsize = 30)
plt.xticks(fontsize = 25, rotation = 70)
plt.yticks(fontsize = 20)
plt.xlabel(" ",fontsize = 30)
plt.ylabel("Número de livros best sellers",fontsize = 25)

plt.show()
```

#### Autores mais Lidos:

```python
#Agrupando por autores que tiveram mais resenhas:
authors_by_reviews = amazon.groupby('Author')[['Reviews']].mean()\
                                            .sort_values('Reviews', ascending=False)\
                                            .reset_index()
#authors_by_reviews
```

```python
# top 10 autores mais lidos:
plt.figure(figsize = (30,10))

sns.barplot(
    x = authors_by_reviews["Author"][:10], 
    y = authors_by_reviews["Reviews"][:10],
    palette = 'magma' )

sns.lineplot(
    x = authors_by_reviews["Author"][:10], 
    y = authors_by_reviews["Reviews"][:10],
    color = "Salmon" )

plt.title("Top 10 Autores Best Sellers", fontsize = 30)
plt.xticks(fontsize = 25, rotation = 70)
plt.yticks(fontsize = 20)
plt.xlabel(" ",fontsize = 30)
plt.ylabel("Número de reviews",fontsize = 25)

plt.show()
```

#### Livros Mais Lidos:

```python
# Agrupamos os livros por número de resenhas:
books_by_reviews = amazon.groupby('Name')[['Reviews']].sum()\
                                                .sort_values('Reviews', ascending=False)\
                                                .reset_index()
#books_by_reviews
```

```python
# top 5 livros mais lidos:
plt.figure(figsize = (30,15))

sns.barplot(
    x = books_by_reviews["Name"][:5], 
    y = books_by_reviews["Reviews"][:5],
    palette = 'magma' )

sns.lineplot(
    x = books_by_reviews["Name"][:5], 
    y = books_by_reviews["Reviews"][:5],
    color = "Salmon" )

plt.title("Top 5 Livros mais Lidos", fontsize = 30)
plt.xticks(fontsize = 25, rotation = 70)
plt.yticks(fontsize = 20)
plt.xlabel(" ",fontsize = 25)
plt.ylabel("Número de reviews",fontsize = 25)

plt.show()
```

#### Livros por Gênero:

```python
# Agrupamos o número de livros por gênero:
books_by_genre = amazon.groupby('Genre')[['Name']].count()\
                                         .sort_values('Name', ascending=False)\
                                         .reset_index()
books_by_genre
```

```python
# Gêneros dos livros best sellers:
plt.figure(figsize=(7,5))

sns.barplot(
    x = books_by_genre["Genre"][:2], 
    y = books_by_genre["Name"][:2],
    palette = 'magma' )

plt.title("Gêneros dos Livros Beste Sellers")
plt.xticks(rotation = 70)
plt.yticks()
plt.xlabel(" ")
plt.ylabel("Número de livros")

plt.show()
```

```python
# Definindo os parâmetros de visualização:
fig_2 = make_subplots(rows=1, cols=1,
                      specs=[[{'type': 'domain'}]],
                      subplot_titles=(""))

# Configurando os parâmetros do gráfico de pizza:
fig_2.add_trace(go.Pie(values=books_by_genre['Name'],
                       title='Gênero',
                       labels=['Não Ficção', 'Ficção'],
                       textinfo='label+percent',
                       insidetextorientation='radial',
                       hole=0.3,
                       marker=dict(colors=['Gray','Salmon'], 
                                   line_color='Gray', 
                                   line_width=1),
                       textfont={'color': '#FFFFFF', 'size': 12},
                       textfont_size=12))

# Definindo os parâmetros do gráfico ao exibir
fig_2.update_layout(margin_t=30, 
                    showlegend=False, 
                    title_text='Quantidade de Livros por Gênero', 
                    title_x=0.5, 
                    font=dict(family='Arial', 
                              size=12, 
                              color='black'))

# Displaying the graph
fig_2.show()
```

```python
# Definindo os parâmetros de visualização:
fig_3 = make_subplots(rows=1, cols=2,
                      specs=[[{'type': 'xy'}, {"type": "xy"}]],
                      subplot_titles=("Top 10 autores por número de best-sellers", 
                                      "Top 5 livros por número de reviews"))

# Configuração dos parâmetros do gráfico de barra (esquerda):
fig_3.add_trace(go.Bar(x=authors_by_books_written['Author'][:10], 
                       y=authors_by_books_written['Name'][:10],
                       name ='Count books',
                       marker_color=['#ff6666','#f76e6e', '#f07575', '#e87d7d', '#e08585',
                                     '#d98c8c', '#d19494', '#c99c9c', '#c2a3a3', '#baabab']),
                       row=1, col=1)

# Configuração dos parâmetros do gráfico de barra (direita):
fig_3.add_trace(go.Bar(x=books_by_reviews['Name'][:5], 
                       y=books_by_reviews['Reviews'][:5],
                       marker_color=['#ff6666','#f47171', '#e87d7d', '#dd8888', '#d19494'],
                       name ='Reviews'),
                       row=1, col=2)

# Definindo os parâmetros do gráfico ao exibir:
fig_3.update_traces(marker_line_color='Gray',
                    marker_line_width=1)

# Definindo os parâmetros do gráfico ao exibir:
fig_3.update_layout(showlegend=False, 
                    plot_bgcolor='rgba(0,0,0,0)',
                    font=dict(family='Arial', 
                              size=12, 
                              color='black'))

# Plotando os gráficos:
fig_3.show()
```

## 6. Análise Exploratória - Dados Numéricos

Nesta seção, vamos analisar os dados numéricos buscando entender suas relações com os gêneros de ficção e não-ficção.

1. Vamos redenrizar um gráfico de caixa para cada variável numérica.

3. Reperiremos as análises para os gráficos de dispersão e histograma. 
4. Finalizamos esta parte da análise com o gráfico ecdf (Empirical Cumulative Distribution Function ), onde o eixo-x consite no valor da variável em ordem crescente, e o eixo-y é o percentil.

#### Função de distribuição cumulativa empírica:

Fornece uma maneira de modelar a probabilidade cumulativa dos dados da amostra e nos ajuda a estimar a função de distribuição cumulativa.

```python
# Visualização ecdf das avaliações por gênero:
sns.ecdfplot(data=amazon,
             x='User Rating', hue='Genre', 
             palette='magma');
```

No eixo-x temos os valores das avaliações dos usuários, ao passo que o eixo-y vai de 0 até 100%. 

Podemos observar que:

- Menos de 20% do dataset tem avaliações de 4.0 para livros de ficção
- As avaliações de não-ficção são acima de 4.0 estrelas.
- Acima de 40% do dataset as avaliações mais altas (entre 4.6 e 4.8) são para o gênero de ficção.

As classificações do usuário tendem a mostrar uma ligeira correlação negativa com o preço do livro mais vendido.
A maioria dos livros estão na faixa de preço menor que 20 dólares, entretanto não é possível concluir o motivo de alguns livros custarem muito acima ou abaixo do valor médio.


- Relação entre preço e resenhas:

As resenhas dos usuários e o preço também tendem a mostrar uma correlação quase semelhante à classificação e o preço. O que podemos concluir é que tratan-se de variáveis inversamente correlacionadas, ou seja, quanto menor o preço do livro temos mais resenhas e para livros muito mais caros temos poucas resenhas. 

## 8. Considerações Finais

Com base na matriz de correlação construída, bem como nas visualizações apresentadas e discutidas anteriormente, podemos notar que os dados não têm nenhuma relação linear, tanto positiva quanto negativa, entre a classificação dos usuários, as resenhas e o preço dos livros.

```python
# Agrupando duas amostras:
non_fiction = amazon[amazon['Genre'] == 'Non Fiction']['User Rating']
fiction = amazon[amazon['Genre'] == 'Fiction']['User Rating']
```

```python
print(f'Mediana de Não-Ficção: {non_fiction.median()} \nMediana de Ficção: {fiction.median()}')
```

```python
# Visualizando a densidade de distribuição das amostras:
fig_7 = ff.create_distplot([non_fiction, fiction], 
                         ['Non_fiction', 'Fiction'], 
                         colors=['salmon', 'gray',],
                         bin_size=.1, 
                         show_rug=False)

# Adicionando um título:
fig_7.update_layout(title_text='Comparação da Densidade de Distribuição', 
                     template='simple_white',
                     font=dict(family='Arial', size=12, color='black'))
fig_7.show()
```

Como resultado do teste, diferenças estatisticamente significativas foram obtidas entre os grupos comparados. Com base nos resultados obtidos, pode-se argumentar que os usuários avaliam os livros de maneira diferente dependendo do gênero e, com base nos valores medianos, os leitores dão preferência às obras de ficção.

No decorrer da análise, estabelecemos quais autores receberam as avaliações mais altas dos leitores, quais autores escreveram mais best-sellers, quais livros receberam mais resenhas dos leitores. Além disso, verificamos que a literatura de não ficção está se tornando cada vez mais um best-seller, mas os usuários também gostam mais de ficção, o que é confirmado pelos resultados estatisticamente significativos obtidos durante o teste final.
